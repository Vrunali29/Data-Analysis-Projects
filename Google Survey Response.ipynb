```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
 
# Load the CSV file
df = pd.read_csv("Survey on Generative AI for Data Preprocessing.csv")
 
# Strip any leading or trailing spaces from column names
df.columns = df.columns.str.strip()
 
# Set style for better visuals
sns.set(style="whitegrid")
 
# Figure 1: Total number of participants by role
plt.figure(figsize=(10, 5))
sns.countplot(y=df["What is your primary role in data science?"], order=df["What is your primary role in data science?"].value_counts().index)
plt.title("Total Number of Participants by Role")
plt.xlabel("Count")
plt.ylabel("Role in Data Science")
plt.show()
 
# Figure 2: Role by years of experience
plt.figure(figsize=(12, 6))
sns.countplot(y=df["How many years of experience do you have in data preprocessing?"], 
              hue=df["What is your primary role in data science?"], 
              palette="coolwarm")
plt.title("Role by Years of Experience")
plt.xlabel("Count")
plt.ylabel("Years of Experience")
plt.legend(title="Role in Data Science")
plt.show()
 
 
# Figure 3: AI tool usage
plt.figure(figsize=(8, 4))
sns.countplot(x=df["If you use generative AI tools, which ones do you primarily use?"], hue=df["If you use generative AI tools, which ones do you primarily use?"], palette="pastel", legend=False)
plt.title("Usage of Generative AI Tools for Data Preprocessing")
plt.xlabel("Used Generative AI Tools")
plt.ylabel("Count")
plt.show()
 
# Figure 4: Perceived Efficiency Improvement
plt.figure(figsize=(10, 5))
sns.countplot(y=df["Have you noticed an improvement in preprocessing efficiency since using generative AI tools?"], order=df["Have you noticed an improvement in preprocessing efficiency since using generative AI tools?"].value_counts().index)
plt.title("Perceived Improvement in Preprocessing Efficiency")
plt.xlabel("Count")
plt.ylabel("Response")
plt.show()
 
# Figure 5: Reduction in Manual Effort
plt.figure(figsize=(10, 5))
sns.countplot(y=df["Do generative AI tools reduce manual effort in preprocessing tasks?"], order=df["Do generative AI tools reduce manual effort in preprocessing tasks?"].value_counts().index)
plt.title("Reduction in Manual Effort")
plt.xlabel("Count")
plt.ylabel("Response")
plt.show()
 
# Figure 6: Trust in AI-generated results
plt.figure(figsize=(10, 5))
sns.countplot(y=df["Do you trust the preprocessing results generated by AI tools without manual verification?"], order=df["Do you trust the preprocessing results generated by AI tools without manual verification?"].value_counts().index)
plt.title("Trust in AI-Generated Preprocessing Results")
plt.xlabel("Count")
plt.ylabel("Response")
plt.show()
 
 
# Figure 8: Willingness to Recommend AI Tools
plt.figure(figsize=(10, 5))
sns.countplot(y=df["Would you recommend generative AI tools for data preprocessing to other data scientists?"], order=df["Would you recommend generative AI tools for data preprocessing to other data scientists?"].value_counts().index)
plt.title("Would You Recommend AI Tools for Data Preprocessing?")
plt.xlabel("Count")
plt.ylabel("Response")
plt.show()
```
